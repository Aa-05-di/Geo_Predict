```python
from dotenv import load_dotenv
import os
load_dotenv()
key=os.getenv("hug_key")
```

```python
from langchain_huggingface import HuggingFaceEndpoint

repo_id = "mistralai/Mistral-7B-Instruct-v0.3"
llm = HuggingFaceEndpoint(
    repo_id=repo_id,
    task="text-generation",
    max_length=128,
    temperature=0.6,
    token=key,
    timeout=60 
)

prompt="Trump's iPhone olive branch is a significant trade war retreat Sentiment Score: -0.3182.Is the following news article likely to influence stock markets?If yes,state yes and explain how. "

# Now test it
response = llm.invoke(prompt)
print(response)
```
